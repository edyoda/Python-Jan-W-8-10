### Insert 
> - insert_one()
> - insert_many()
> - insert()

```python
collection.insert_one({'id':101,'name':"ABC",'email':"abc@xyz.com",'salary':45000})
```

```python
collection.insert_many([{'id':102,'name':"PQR",'email':"pqr@xyz.com"},{'id':103,'name':"RST",'email':"rst@xyz.com",'contact':'7123456789'}])
```

### Find
> - find()
> - find_one()

```python
g = collection.find()
print(g)
```

```python
for obj in g:
    print(obj)
```
